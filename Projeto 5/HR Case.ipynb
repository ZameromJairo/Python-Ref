```python
# Modeling Librarys
import numpy as np
import pandas as pd

# DataViz Librarys
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.shape
```

```python
data.columns
```

### Missing Data

```python
data.isnull().sum()
```

```python
# No data is missing....
sns.heatmap(data.isnull(),cbar = False)
```

### Exploratory Analysis

```python
data.describe()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(7,6))
plt.title('Salary Analysis')
sns.kdeplot(data['Salary'],color = 'Red')
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(7,6))
plt.title('Experience Analysis')
sns.kdeplot(data['YearsExperience'],color = 'Blue')
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize=(10,5))
sns.boxenplot(data['Salary'])
plt.show()
```

```python
sns.set_style("ticks")
plt.figure(figsize=(10,5))
sns.boxenplot(data['YearsExperience'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x = 'Salary', y = 'YearsExperience')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=data, x = 'Salary', y = 'YearsExperience')
plt.show()
```

```python
correlacao_pearson = data.corr('pearson')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao_pearson,annot=True,cbar = False)
```

```python
correlacao_spearman = data.corr('spearman')
plt.figure(figsize=(10,5))
sns.heatmap(correlacao_spearman,annot=True,cbar = False)
```
