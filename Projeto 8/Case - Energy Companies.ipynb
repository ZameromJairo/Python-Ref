```python
# Modeling Librarys
import numpy as np
import pandas as pd

# DataViz Librarys
import matplotlib.pyplot as plt
import seaborn as sns
```

### Exploratory Analysis

```python
import random
def random_color():
    levels = range(32,256,32)
    return tuple(random.choice(levels) for _ in range(3))
```

```python
# graphic

# Style
plt.style.use('seaborn-darkgrid')

# Size
plt.figure(figsize=(20,8))

# Title
plt.title('Brazilian Energy Companies Analysis', fontsize = 20, weight = 'bold')

# Plot Petrobras
plt.plot(data.index,data.Petrobras, color = '#008c4a', linewidth = 2)

# Petrobras text
plt.text(data.index[-1],data['Petrobras'].tail(1), 'Petrobras', color = '#008c4a', size = 'large', horizontalalignment = 'left')

# labels
plt.xlabel('Period', fontsize = 14)
plt.ylabel('Closing Price (R$)', fontsize = 14)

# All columns plot
for column in data.columns[1:]:
    # Plot all columns
    plt.plot(data.index,data[column], color = 'gray', linewidth = 1)
    plt.text(data.index[-1],data[column].tail(1), column, color = 'gray', size = 'large', horizontalalignment = 'left')


plt.show()
```
